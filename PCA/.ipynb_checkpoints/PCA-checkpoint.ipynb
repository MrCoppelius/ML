```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA

data = pd.read_csv('close_prices.csv')
djia_index =  pd.read_csv('djia_index.csv')
```

```python
data.corr(method='pearson')
```

```python
pca = PCA(n_components=10)
pca.fit(data.values[:,1:])
```

```python
X = pca.transform(data.values[:,1:])
```

```python
pca.explained_variance_ratio_
```

```python
np.corrcoef( [X[:,0].astype(float), djia_index.values[:,1].astype(float)])
```
