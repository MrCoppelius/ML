```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble  import GradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
```

```python
data = pd.read_csv('gbm-data.csv')
```

```python
X = data.iloc[:,1:].values
y = data.iloc[:,0].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.8, random_state=241)
```

```python
learning_rate_list = [0.2] #[1, 0.5, 0.3, 0.2, 0.1]
loss_test = dict.fromkeys(learning_rate_list, [])
loss_train = dict.fromkeys(learning_rate_list, [])
for learning_rate in learning_rate_list: 
    clf = GradientBoostingClassifier(n_estimators = 250,learning_rate = learning_rate, random_state=241)
    clf.fit(X_train, y_train)
    y_pred_test = clf.staged_predict_proba(X_test)
    y_pred_train = clf.staged_predict_proba(X_train)
    for  y_pred_test_it, y_pred_train_it in zip(y_pred_test,y_pred_train):
        loss_test[learning_rate].append(log_loss(y_test, y_pred_test_it))
        loss_train[learning_rate].append(log_loss(y_train, y_pred_train_it))
```

```python
clf_RF = RandomForestClassifier(n_estimators = 200, random_state=241)
clf_RF.fit(X_train, y_train)
y_pred_test_RF = clf_RF.predict_proba(X_test)
y_pred_train_RF = clf_RF.predict_proba(X_train)
loss_test_RF = log_loss(y_test, y_pred_test_RF)
loss_train_RF = log_loss(y_train, y_pred_train_RF)
```
